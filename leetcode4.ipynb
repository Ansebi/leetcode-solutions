# Binary Search

```python
import numpy as np
```

```python
sorted_array = list(np.arange(1000) * 5)
```

with the means of binary search, let's find where an arbitrary element belongs in the array

```python
def binary_search(sorted_array, value, idx_min=None, idx_max=None, first_run=True):
    'returns idx of the value'
    if first_run:
        if not sorted_array:
            return None
        if value < sorted_array[0]:
            return -0.5
        if value == sorted_array[0]:
            return 0
        if value > sorted_array[-1]:
            return len(sorted_array) - 0.5
        if value == sorted_array[-1]:
            return len(sorted_array) - 1
        if idx_min is None:
            idx_min = 0
        if idx_max is None:
            idx_max = len(sorted_array)
    idx_middle = idx_min + (idx_max - idx_min) // 2
    # print(idx_min, idx_middle, idx_max)
    if idx_min == idx_middle or idx_middle == idx_max:
        return idx_middle + 0.5
    element_middle = sorted_array[int(idx_middle)]
    if value == element_middle:
        return idx_middle
    if value > element_middle:
        idx_min = idx_middle
    elif value < element_middle:
        idx_max = idx_middle
    return binary_search(sorted_array, value, idx_min, idx_max, first_run=False)


def pos_and_median(sorted_array):
    odd = False
    len_ = len(sorted_array)
    if len_ % 2:
        odd = True
    if odd:
        pos_median = len_ // 2
        median = sorted_array[pos_median]
    else:
        pos_median_right = len_ // 2
        pos_median = pos_median_right - 0.5
        pos_median_left = pos_median_right - 1        
        median = (sorted_array[pos_median_left] + sorted_array[pos_median_right]) / 2
    return pos_median, median
```
